```python
from sklearn.metrics import f1_score, classification_report,confusion_matrix
import numpy as np
import seaborn as sn
import pandas as pd
import os
from matplotlib import pyplot as plt
```

```python
def get_report(path):
    dev_result = pd.read_csv(os.path.join(path,'dev_sub.csv'))
    ground_truth = dev_result.label
    pre = dev_result.label_pre
    f1 = f1_score(ground_truth,pre,average="macro")
    report = classification_report(ground_truth,pre)
    cm = confusion_matrix(ground_truth, pre)
    return f1,report,cm
def file_report(path):
    dev_result = pd.read_csv(path)
    ground_truth = dev_result.label
    pre = dev_result.label_pre
    f1 = f1_score(ground_truth,pre,average="macro")
    report = classification_report(ground_truth,pre)
    cm = confusion_matrix(ground_truth, pre)
    return f1,report,cm
```

```python
fig,ax = plt.subplots()
sn.heatmap(cm,cmap="GnBu",ax=ax)
ax.set_xlabel("prelabel")
ax.set_ylabel("truelabel")
```
