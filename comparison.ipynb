```python
import numpy as np
import pandas as pd
import os
```

```python
test_file = pd.read_csv("./data/nlp_new_train0/test.csv")
```

```python
test_file.head()
```

```python
def file_name(file_dir):  
    for root, dirs, files in os.walk(file_dir): 
        return files #当前路径下所有非目录子文件
file_name = file_name("./input_big/")
file_name = ["./submit/90.06.csv"]
real_file = pd.read_csv("./submit/adv_bert_89.9.csv",sep="\t",header=None,names=['id','label'])
real_file = pd.merge(test_file[['id','question1',"question2"]],real_file,on='id')
```

```python
disagree =[]
for f_n in file_name :
    file_small = pd.read_csv(f_n,sep="\t",header=None,names=['id','label'])
    file_small.sort_values("id",inplace=True)
#     print(f_n)
    sum=0
    index1=0
    for index,row in file_small.iterrows():
        
        if(real_file['label'][index1]==row['label']):
            sum=sum+1
        else:
            disagree.append(index)
        index1=index1+1
        
        
        

    print(sum/len(file_small['label']),"---相似度",f_n)
```
